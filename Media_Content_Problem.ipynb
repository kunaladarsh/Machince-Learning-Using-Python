# To Identify the reason for reduction in viewership of a show telecasted by a media company.
> Data clean/prepare & derive new feature from existing

> Bulid and train Model to identify the reason for reduction in viewership

> To boost the preformance of Model

**First, import the required modules**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
```

**Load Dataset to pandas data frame**

```python
df = pd.read_csv("mediacompany.csv")
df = df.drop('Unnamed: 7', axis = 1)
df.head()
```

**Data Preparation and derive new feature from existing features.**

```python
# Converting date to pandas detetime format

df['Date'] = pd.to_datetime(df['Date'])

# Extract day of week from date

df['Date'] = pd.to_datetime(df['Date'])
df['day_of_week'] = df['Date'].dt.dayofweek 
df.head()
```

```python
# create a weekend feature with value 1 for weekends and 0 for weekdays

li =[]
for i in (df.day_of_week):
    if (i == 5) or (i==6):
        li.append(1)
    else:
        li.append(0)
    

df['weekend'] = li
df
```

**Seperate target variable**

```python
# putting response variable to x

X = df[['weekend', 'Character_A', 'Ad_impression']]

# putting response variable to y

y = df['Views_show']
```

```python
print(X)
print(y)
```

**Model Building**

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X,y)
```

**Predict**

```python
import statsmodels.api as sm
#Unlike SKLearn, statsmodels don't automatically fit a constant, 
#so you need to use the method sm.add_constant(X) in order to add a constant. 
X = sm.add_constant(X)
# create a fitted model in one line
lm_1 = sm.OLS(y,X).fit()
print(lm_1.summary())
```
